```python
from bs4 import BeautifulSoup
import requests as rq
import pandas as pd
```

Tabela na Wikipedia

```python
url = "https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Pernambuco"
headers = {'User-Agent':'Mozilla/5.0'}
page = rq.get(url, headers = headers)
parser = BeautifulSoup(page.text,'html.parser')
```

```python
table=[]
linha = parser.find('table', class_="wikitable sortable").find_all("tr")
for x in range(1,186):
    aux_table=[]
    coluna = linha[x].find_all('td')
    aux_table.append(coluna[1].text)
    aux_table.append(coluna[4].text)
    table.append(aux_table)
    
pernambuco = pd.DataFrame (table, columns=["Municipio", "Microrregiao"])
```

Correção dos dados (dado em falta)

```python
pernambuco.loc[pernambuco['Municipio'] == 'Jupi', 'Microrregiao'] = 'Garanhuns'
```

Salvo em csv

```python
pernambuco.to_csv('../microrregioes/municipios_microrregioes_pernambuco.csv', index=False)
```
